## Shakespeare Modellieren

### Shakespeare Corpus in NLTK

```python
import nltk
nltk.download('shakespeare')
from nltk.corpus import shakespeare
from xml.etree import ElementTree
shakespeare.fileids()
```

```python
import re 
from nltk.tokenize import word_tokenize
nltk.download('punkt')

shakespeare_corpus = ""
for play_title in shakespeare.fileids():
    play = shakespeare.xml(play_title)
    
    for line in play.findall('ACT/SCENE/SPEECH/LINE'):
        if line.text is not None:
            #print(line.text)
            shakespeare_corpus += line.text
            shakespeare_corpus += '\n'

print(shakespeare_corpus[0:200])
```

```python
shakespeare_corpus = word_tokenize(shakespeare_corpus)
print(shakespeare_corpus[0:100])
```

### Endlos würfelnde Affen: Shakespeare mit 0-Grammen

```python
shakespeare_voc = list(set(shakespeare_corpus))
shakespeare_voc_size = len(shakespeare_voc)
shakespeare_voc_size
```

P(w) = 1/voc_size

```python
P_w = 1/shakespeare_voc_size
print("P(w)  = " + '{:f}'.format(P_w)) # no scientific notation
```

```python
import random

def generate_model_0(voc, n_tokens):
    for i in range(n_tokens):
        word = random.choice(voc)
        print(word, end=' ')
```

```python
generate_model_0(shakespeare_voc,30)
```

### Würfelnde Affen (mit gezinkten Würfeln): Shakespeare mit 1-Grammen

```python
shakespeare_fd = nltk.FreqDist(shakespeare_corpus)
shakespeare_fd.most_common()
```

```python
shakespeare_corpus_size = len(shakespeare_corpus)
shakespeare_corpus_size
```

```python
print("'the' frequency: " + str(shakespeare_fd['the']))
print("'the' P: " + str(shakespeare_fd['the']/shakespeare_corpus_size))
```

```python
shakespeare_pd = nltk.MLEProbDist(shakespeare_fd)
shakespeare_pd.prob('the')
```

```python
def generate_model_1(pd, n_tokens):
    for i in range(n_tokens):
        word = pd.generate()
        print(word, end=' ')
```

```python
generate_model_1(shakespeare_pd,30)
```

### Shakespeare mit Bigrammen

```python
from nltk import bigrams

shakespeare_bigrams = nltk.bigrams(shakespeare_corpus)
shakespeare_cfd = nltk.ConditionalFreqDist(shakespeare_bigrams) # conditional frequency distribution
```

```python
shakespeare_cfd['My']
```

```python
def generate_model2_max(cfdist, word, n_tokens):
    for i in range(n_tokens):
        print(word, end=' ')
        word = cfdist[word].max()
```

```python
generate_model2_max(shakespeare_cfd, 'Once', 100)
```

```python
def generate_model2_rand(cfdist, word, n_tokens):
    for i in range(n_tokens):
        print(word, end=' ')
        words = list(cfdist[word])
        word = random.choice(words)
```

```python
generate_model2_rand(shakespeare_cfd, 'Once', 100)
```

```python
generate_model2_rand(shakespeare_cfd, 'What', 100)
```

```python
generate_model2_rand(shakespeare_cfd, 'She', 100)
```

```python
def generate_model2_loaded_dice(cfdist, word, n_tokens):
    for i in range(n_tokens):
        print(word, end=' ')
        pd = nltk.MLEProbDist.cfdist[word]
        word = pd.generate()
```

```python
generate_model2_rand(shakespeare_cfd, 'She', 100)
```

### Skahespeare mit Trigrammen modellieren

```python
shakespeare_trigrams = nltk.trigrams(shakespeare_corpus)
condition_pairs = (((w0, w1), w2) for w0, w1, w2 in shakespeare_trigrams)
shakespeare_cfd = nltk.ConditionalFreqDist(condition_pairs)
```

```python
def generate_model3(cfdist, w1, w2, n_tokens):
    for i in range(n_tokens):
        print(w1, end=' ')
        w3 = cfdist[w1, w2].max()
        w1 = w2
        w2 = w3
```

```python
generate_model3(shakespeare_cfd, 'when', 'they', 100)
```

```python
def generate_model3_loaded_dice(cfdist, w1, w2, n_tokens):
    for i in range(n_tokens):
        print(w1, end=' ')
        cfd = cfdist[w1, w2]
        pd = nltk.MLEProbDist(cfd)
        w3 = pd.generate()  
        w1 = w2
        w2 = w3
```

```python
generate_model3_loaded_dice(shakespeare_cfd, 'when', 'they', 100)
```

### Skahespeare mit 4-grammen modellieren

```python
from nltk import ngrams
shakespeare_4grams = ngrams(shakespeare_corpus,4)
condition_pairs = (((w0, w1, w2), w3) for w0, w1, w2, w3 in shakespeare_4grams)
shakespeare_cfd = nltk.ConditionalFreqDist(condition_pairs)
```

```python
def generate_model4(cfdist, w1, w2, w3, n_tokens):
    for i in range(n_tokens):
        print(w1, end=' ')
        w4 = cfdist[w1, w2, w3].max()
        w1 = w2
        w2 = w3
        w3 = w4
```

```python
generate_model4(shakespeare_cfd, 'When', 'they', 'are', 100)
```

```python
def generate_model4_loaded_dice(cfdist, w1, w2, w3, n_tokens):
    for i in range(n_tokens):
        print(w1, end=' ')
        cfd = cfdist[w1, w2, w3]
        pd = nltk.MLEProbDist(cfd)
        w4 = pd.generate()  
        w1 = w2
        w2 = w3
        w3 = w4
```

```python
generate_model4_loaded_dice(shakespeare_cfd, 'When', 'they', 'are', 100)
```
